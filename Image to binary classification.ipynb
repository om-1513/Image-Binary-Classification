```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
!rmdir /content/dataset/training_set/.ipynb_checkpoints/
!rmdir /content/dataset/test_set/.ipynb_checkpoints
!rmdir /content/dataset/.ipynb_checkpoints
!rmdir /content/.ipynb_checkpoints
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255, 
                                   shear_range = 0.2,  
                                   zoom_range = 0.2,   
                                   horizontal_flip = True) 
training_set = train_datagen.flow_from_directory('/content/dataset/training_set',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
validation_generator = test_datagen.flow_from_directory('/content/dataset/test_set',
                                            target_size = (64, 64), 
                                            batch_size = 32,
                                            class_mode = 'binary')
```

```python

cnn = tf.keras.models.Sequential()
     
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu' , input_shape=[64, 64, 3]))
     
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size= 2, strides= 2))
```

```python

cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size= 2, strides= 2))
```

```python

cnn.add(tf.keras.layers.Flatten())
```

```python

cnn.add(tf.keras.layers.Dense(units= 128, activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units= 1, activation='sigmoid'))
```

```python
cnn.compile(optimizer='adam', loss='binary_crossentropy',  metrics= ['accuracy'])
```

```python
cnn.fit(x = training_set, validation_data= training_set, epochs= 25)
```
